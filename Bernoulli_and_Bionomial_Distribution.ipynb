```python
import scipy
from scipy.stats import bernoulli
import seaborn as sns
sns.set(color_codes=True)
```

```python
bernolli_rvs=bernoulli.rvs(p=0.6,size=1000)
```

```python
ax=sns.displot(bernolli_rvs,bins=100,kde=False)
ax.set_xlabels("Bernoulli Random Variables")
ax.set_ylabels("Frequency")
```

```python
import numpy as np
```

```python
bionom_rvs = np.random.binomial(n=10, p=0.6, size=100)
```

```python
ax=sns.displot(bionom_rvs,bins=20,kde=False)
ax.set_xlabels("Bionomial Distribution")
ax.set_ylabels("Frequency")
```
