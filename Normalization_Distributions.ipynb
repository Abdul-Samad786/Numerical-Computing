```python
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
data_normal=norm.rvs(size=1000,loc=0,scale=1)
```

```python
ax=sns.displot(
    data_normal,
    bins=50,
    kde=True
)
plt.xlim(-10,10)
ax.set(xlabel='Standard Distribution',ylabel='Frequency')
```

```python
Norm_var=norm.rvs(size=1000,loc=2,scale=3)
bx=sns.displot(
    Norm_var,
    bins=50,
    kde=True)
plt.xlim(-10,10)
bx.set(xlabel='Normal Distribution',ylabel='Frequency')
```
